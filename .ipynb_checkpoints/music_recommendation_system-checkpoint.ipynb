```python
import numpy as np
import pandas as pd
import re
```

<!-- DATA Cleaning -->

```python
music_dt = music_dt.drop('User-Rating', axis=1)
```

```python
music_dt
```

```python
music_dt = music_dt.rename(columns={'Song-Name': 'SongName', 'Singer/Artists': 'Artists', 'Album/Movie': 'Movie' })
```

```python
music_dt
```

```python
# Remove the null values
print(music_dt.isnull().sum())
```

```python
music_dt.dropna(inplace=True)
# music_cln = music_dt.dropna()
# music_cln
```

```python
print(music_dt.isnull().sum())
```

```python
# Remove duplicate data
music_dt.duplicated().sum()
```

```python
music_dt = music_dt.drop_duplicates()
music_dt
```

```python
# Data Preprocessing
music_dt.iloc[0].Genre
```

```python
def split_genre(text):
    return re.sub(r'([a-z])([A-Z])', r'\1, \2', text)  # Inserts commas before uppercase letters

# Apply the function safely using .loc[]
music_dt.loc[:, 'Genre'] = music_dt['Genre'].apply(split_genre)
music_dt.loc[:, 'Artists'] = music_dt['Artists'].apply(split_genre)

# Display results
print(music_dt.head())
```

```python
from IPython.display import display
display(music_dt)
```

```python
# # music_dt.loc[:, 'Artists'] = music_dt['Artists'].apply(lambda x: "".join(x.split(" ")))
# music_dt.loc[:, 'Artists'] = music_dt['Artists'].apply(lambda x:[i.replace(" ",""))
music_dt.loc[:, 'Artists'] = music_dt['Artists'].apply(lambda x: x.replace(" ", "").replace(",", ", "))

# music_dt['Artists'] = music_dt['Artists'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
music_dt.head()
```

```python
music_dt['Tags'] = music_dt['Artists'] + ", " + music_dt['Genre'] + ", " + music_dt['Movie']
```

```python
music_dt.head()
```

```python
new_music = music_dt[['SongName','Tags']] 
```

```python
new_music
```

```python
new_music['Tags'] = new_music['Tags'].apply(lambda x:x.lower().replace(",", ""))
```

```python
new_music.head()
```

```python
# import nltk
# from nltk.stem.porter import PorterStemmer

# ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vectors = cv.fit_transform(new_music['Tags']).toarray()
```

```python
vectors.shape
```

```python
feature_names = cv.get_feature_names_out()
print(list(feature_names))
```

```python
# from sklearn.metrics.pairwise import cosine_similarity
# similarity = cosine_similarity[vectors]

from sklearn.metrics.pairwise import cosine_similarity

# Example: Compute similarity between vectors
similarity = cosine_similarity(vectors)  # âœ… Use parentheses, not square brackets

# print(similarity)
```

```python
def recommendation(SongName):
    idx = new_music[new_music['SongName'] == SongName].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for i in distances[1:6]:
        songs.append(new_music.iloc[i[0]].SongName)

    return songs
```

```python
recommendation('Coca Cola')
```

```python
import pickle
pickle.dump(new_music,open('new_music.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
